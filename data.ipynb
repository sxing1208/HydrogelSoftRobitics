```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
value_1 = []
value_2 = []
value_3 = []
value_4 = []
for line in open('nrf52 cycles 1.txt'):
    value_1.append(line[64])
    value_2.append(line[65])
    value_3.append(line[67])
    value_4.append(line[68])

print(value_1)
print(value_2)
print(value_3)
print(value_4)
```

```python
value_2_1 = []
value_2_2 = []
value_2_3 = []
value_2_4 = []
for line in open('nrf52 cycle 2.txt'):
    value_2_1.append(line[64])
    value_2_2.append(line[65])
    value_2_3.append(line[67])
    value_2_4.append(line[68])

num_2 = []
for index in np.arange(len(value_2_1)):
    num_2_1 = dict_[value_2_1[index]]*16
    num_2_2 = dict_[value_2_2[index]]
    num_2_3 = dict_[value_2_3[index]]*16*16*16
    num_2_4 = dict_[value_2_4[index]]*16*16
    num_2.append(num_2_1+num_2_2+num_2_3+num_2_4)
    #print(num_2_1+num_2_2+num_2_3+num_2_4)
print(num_2)
plt.plot(num_2)
```

```python
value_3_1 = []
value_3_2 = []
value_3_3 = []
value_3_4 = []
for line in open('nrf52 cycles 3.txt'):
    #print(line)
    value_3_1.append(line[64])
    value_3_2.append(line[65])
    value_3_3.append(line[67])
    value_3_4.append(line[68])

num_3 = []
for index in np.arange(len(value_2_1)):
#for index in [0,1]:
    num_3_1 = dict_[value_3_1[index]]*16
    num_3_2 = dict_[value_3_2[index]]
    num_3_3 = dict_[value_3_3[index]]*16*16*16
    num_3_4 = dict_[value_3_4[index]]*16*16
    num_3.append(num_3_1+num_3_2+num_3_3+num_3_4)
    print(num_3[index])
#print(num_3)
print(len(num_3))
plt.plot(num_3)
```

```python
line_count = 0
main_list_sensor = []
main_list_stimulation = []
for line in open('cycle 3.csv'):
    if line_count == 0:
        line_count+=1
    else:
        if line_count == 100:
            line_list = line[:-1].split(',')
            if float(line_list[1])<1:
                main_list_stimulation.append((float(line_list[1])-0.41))
            else:
                main_list_stimulation.append((float(line_list[1])+0.31))
            main_list_sensor.append(float(line_list[2]))
            line_count = 1
        else:
            line_count+=1
print(len(main_list_sensor))
plt.plot(main_list_sensor[13:670])
plt.plot(main_list_stimulation[13:670])
```

```python
sensor_array = np.array(main_list_sensor[13:670])
stimulation_array = np.array(main_list_stimulation[13:670])
bluetooth_array = np.array(num_3)
import scipy
bluetooth_array = scipy.signal.resample(bluetooth_array, len(sensor_array)+1)[:-1]
bluetooth_array = scipy.signal.savgol_filter(bluetooth_array, 51, 2)
with open("output.csv","w") as f:
    for index in np.arange(len(sensor_array)):
        f.write(str(sensor_array[index])+','+str(stimulation_array[index])+','+str(bluetooth_array[index])+"\n")

```
